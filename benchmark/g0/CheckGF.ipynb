```python
import numpy as np 
g = np.loadtxt('g0.dat')

from pytriqs.gf.local import *
from pytriqs.utility import mpi
from pytriqs.applications.impurity_solvers.ctint_new import CtintSolver

S = CtintSolver()

S.solve(U = 0.2, beta = 100.0, tmax = 10.0, Gamma = 0.25, alpha = 0.0, Edp = 0,
        n_cycles = 10, n_warmup_cycles = 10)
```

```python
from pytriqs.plot.mpl_interface import oplot, plt
oplot(S.G0_w[0,0], '-o', x_window = (-30,30))
#plt.xlim(-50,50)
#plt.ylim(-2,1)
#S.G0_w.data
```

```python
#oplot(S.G0_t[0,0].imag, '-o', x_window = (-5,5), label = 'Im G ++')
oplot(S.G0_t[0,1].imag, '-o', x_window = (-5,5), label = 'Im G <')
plot(g[:,0], g[:,2], '-o', label = 'Im Gref <')
plot(g[:,0], -(g[:,2] + g[:,4]), '-x', label = ' - Im Gref ++')
plt.legend()
```

```python
oplot(S.G0_t.real, '-o', x_window = (-10,10))
plot(g[:,0], g[:,1], '-o', label = 'Re Gref <')
plot(g[:,0], (g[:,1] + g[:,3]), '-x', label = 'Re Gref ++')
plt.grid()
plt.legend()
```
