```python
from series import *
from qmc_keldysh import *

alpha = 0.5
Gamma = 1.0
#U = 1.0 * Gamma
epsilon_d = 0.01
tmax = 3. * Gamma
V = 2. * Gamma
muL = V/2.
muR = -V/2.
beta = 200.
GF_type = 1 #Werner
is_current = True

length_cycle = 4
n_cycles = 10000
n_warmup_cycles = 1000
Nmax = 8
U_qmc = 4.
p_dbl = 1.
if_ph = True
name='test_Werner.h5'

p = SimpleNamespace(alpha=alpha,
                    epsilon_d=epsilon_d,
                    Gamma=Gamma,
                    tmax=tmax,
                    muL=muL,
                    muR=muR,
                    beta=beta,
                    is_current = is_current,
                    GF_type = GF_type)
p_qmc = SimpleNamespace(U_qmc = U_qmc,
                        n_cycles = n_cycles,
                        n_warmup_cycles = n_warmup_cycles, 
                        length_cycle = length_cycle, 
                        LastOrder=Nmax,
                        p_dbl=p_dbl)
res = QMC(p,p_qmc, True, verbosity=3, if_ph=if_ph)
save(p,p_qmc,res,name)
```
